## Training from scratch

```python
%%bash

gpu_ids=0

dataset_model=iPER_temporal_smoothing
data_dir=/kaggle/input/iperdata/iPER-data
images_folder=images
smpls_folder=smpls
ids_dir=/kaggle/working/impersonator
train_ids_file=train.txt
test_ids_file=val.txt

checkpoints_dir=/kaggle/working
name=exp_iPER_temporal_smoothing

model=impersonator_trainer_temporal_smoothing
gen_name=impersonator_temporal_smoothing
image_size=256

load_path="None"
batch_size=12
lambda_rec=10.0
lambda_tsf=10.0
lambda_face=5.0
lambda_style=0.0
lambda_mask=1.0
lambda_mask_smooth=1.0
nepochs_no_decay=5
nepochs_decay=25

python train_temporal_smoothing.py --gpu_ids ${gpu_ids}        \
    --data_dir  ${data_dir}                 \
    --images_folder    ${images_folder}     \
    --smpls_folder     ${smpls_folder}      \
    --checkpoints_dir  ${checkpoints_dir}   \
    --ids_dir          ${ids_dir}           \
    --train_ids_file   ${train_ids_file}    \
    --test_ids_file    ${test_ids_file}     \
    --load_path        ${load_path}         \
    --model            ${model}             \
    --gen_name         ${gen_name}          \
    --name             ${name}              \
    --dataset_mode     ${dataset_model}     \
    --image_size       ${image_size}        \
    --batch_size       ${batch_size}        \
    --lambda_face      ${lambda_face}       \
    --lambda_tsf       ${lambda_tsf}        \
    --lambda_style     ${lambda_style}      \
    --lambda_rec       ${lambda_rec}         \
    --lambda_mask      ${lambda_mask}       \
    --lambda_mask_smooth  ${lambda_mask_smooth} \
    --nepochs_no_decay ${nepochs_no_decay}  --nepochs_decay ${nepochs_decay}  \
    --mask_bce     --use_vgg       --use_face    --use_temporal_smoothing
```
